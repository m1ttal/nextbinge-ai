```python
def extract_collection_name(value):
    if pd.isna(value):
        return ""

    try:
        data = ast.literal_eval(value)
        return data.get("name", "")
    except:
        return ""
```

```python
dim_items["collection_name"] = (
    dim_items["belongs_to_collection"]
    .apply(extract_collection_name)
)
```

```python
def extract_company_names(value):
    if pd.isna(value):
        return ""

    try:
        data = ast.literal_eval(value)

        return " ".join(
            company["name"]
            for company in data
            if "name" in company
        )

    except:
        return ""
```

```python
dim_items["company_names"] = (
    dim_items["production_companies"]
    .apply(extract_company_names)
)
```

```python
def extract_country_names(value):
    if pd.isna(value):
        return ""

    try:
        data = ast.literal_eval(value)

        return " ".join(
            country["name"]
            for country in data
            if "name" in country
        )

    except:
        return ""
```

```python
dim_items["country_names"] = (
    dim_items["production_countries"]
    .apply(extract_country_names)
)
```

```python
credits_clean = credits_clean.rename(
    columns={"id": "source_item_id"}
)
```

```python
def extract_cast(cast_str, top_n=5):
    try:
        cast_list = ast.literal_eval(cast_str)
        return [x["name"] for x in cast_list[:top_n]]
    except:
        return []
```

```python
credits_clean["cast_list"] = (
    credits_clean["cast"]
    .apply(extract_cast)
)
```

```python
item_cast = (
    credits_clean[
        ["source_item_id", "cast_list"]
    ]
    .explode("cast_list")
    .rename(
        columns={
            "cast_list": "actor"
        }
    )
)
```

```python
def extract_directors(crew_str):
    try:
        crew_list = ast.literal_eval(crew_str)

        return [
            person["name"]
            for person in crew_list
            if person.get("job") == "Director"
        ]
    except:
        return []
```

```python
credits_clean["director_list"] = (
    credits_clean["crew"]
    .apply(extract_directors)
)
```

```python
item_directors = (
    credits_clean[
        ["source_item_id", "director_list"]
    ]
    .explode("director_list")
    .rename(
        columns={
            "director_list": "director"
        }
    )
)
```
