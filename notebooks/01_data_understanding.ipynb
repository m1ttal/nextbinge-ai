```python
from pathlib import Path
import pandas as pd
```

```python
PROJECT_ROOT = Path.cwd().parent

RAW_DATA= PROJECT_ROOT / "data" / "raw"
```

```python
for folder in RAW_DATA.iterdir():
    print("\n" + "=" * 60)
    print(folder.name)

    for file in folder.iterdir():
        print("  ", file.name)
```

```python
movies_metadata = pd.read_csv(
    RAW_DATA / "movies_dataset" / "movies_metadata.csv",
    low_memory=False
)

credits = pd.read_csv(
    RAW_DATA / "movies_dataset" / "credits.csv"
)

keywords = pd.read_csv(
    RAW_DATA / "movies_dataset" / "keywords.csv"
)

links = pd.read_csv(
    RAW_DATA / "movies_dataset" / "links.csv"
)

ratings = pd.read_csv(
    RAW_DATA / "movies_dataset" / "ratings.csv"
)

anime = pd.read_csv(
    RAW_DATA / "anime_dataset" / "anime-dataset-2023.csv"
)

anime_scores = pd.read_csv(
    RAW_DATA / "anime_dataset" / "users-score-2023.csv"
)

anime_users = pd.read_csv(
    RAW_DATA / "anime_dataset" / "users-details-2023.csv"
)

tvshows = pd.read_csv(
    RAW_DATA / "tvshows_dataset" / "TMDB_tv_dataset_v3.csv"
)
```

```python
for name, df in {
    "movies_metadata": movies_metadata,
    "credits": credits,
    "keywords": keywords,
    "links": links,
    "ratings": ratings,
    "anime": anime,
    "anime_scores": anime_scores,
    "anime_users": anime_users,
    "tvshows": tvshows
}.items():

    print(f"\n{name}")
    print("-" * 50)

    print("Shape:", df.shape)
```

```python
movie_features = content_features.merge(
    dim_items[["item_id", "release_date", "runtime", "language", "popularity", "vote_average", "vote_count", "budget", "revenue", "poster_path"]],
    on = "item_id",
    how = "inner"
)
```

```python
movie_features.to_parquet(
    f"{PROJECT_ROOT}/data/processed/movies/movie_features.parquet"
)
```

```python
movie_features["release_year"] = pd.to_datetime(movie_features["release_date"]).dt.year
```

```python
movie_features["release_year"] = movie_features["release_year"].astype("Int64")
```

```python
movie_features["display_title"] = (
    movie_features["title"].fillna("")
    + " ("
    + movie_features["release_year"]
        .fillna(0)
        .astype(int)
        .astype(str)
    + ")"
)
```
